```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew
```

```python
df_gun = pd.read_csv("/content/drive/MyDrive/tourists/Guns incident Data.csv")

print("\nDataFrame - Gun incident:")
print(df_gun.head(5))
```

```python
#Bar graph
sns.countplot(x='Sex', data=df_gun)
plt.xlabel('Sex')
plt.ylabel('Count')
plt.show()
```

```python
#Pie chart
sex_counts = df_gun['Sex'].value_counts()
labels = sex_counts.index
sizes = sex_counts.values

plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightcoral'])
plt.show()
```

```python
#Histogram
plt.hist(df_gun['Age'], bins=20, color='skyblue', edgecolor='black')
plt.title('Histogram of Gun Incidents by Age')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
#DistPlot
sns.distplot(df_gun['Age'].dropna(), hist=True, kde=True, bins=20, color='darkblue')
plt.title('Distribution Plot of Gun Incidents by Age')
plt.xlabel('Age')
plt.ylabel('Density')
plt.show()
```

```python
#Boxplot
sns.boxplot(x='Age', data=df_gun, color='skyblue')
plt.title('Box Plot of Gun Incidents by Age')
plt.xlabel('Age')
plt.show()
```

```python
df_gun.describe()
```

```python
#Scatterplot
top_2000_df = df_gun.head(2000)

plt.figure(figsize=(12, 8))
sns.stripplot(x='Age', y='Race', data=top_1000_df, jitter=True, alpha=0.7, palette='Set2')
plt.title('Scatter Plot of Age vs. Race')
plt.xlabel('Age')
plt.ylabel('Race')
plt.show()
```

```python
#Barplot
plt.figure(figsize=(8, 6))
sns.barplot(x='Sex', y='Age', data=df_gun, palette='Set2')
plt.title('Average Age Based on Sex')
plt.xlabel('Sex')
plt.ylabel('Average Age')
plt.show()
```

```python
#Boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(x='Age', y='Sex', hue='Police involvement', data=df_gun, palette='Set2')
plt.title('Box Plot of Sex Based on Age with Police Involvement')
plt.xlabel('Age')
plt.ylabel('Sex')
plt.legend(title='Police Involvement', loc='upper right', bbox_to_anchor=(1.25, 1))
plt.show()
```

```python
#Distplot
plt.figure(figsize=(10, 6))
sns.kdeplot(df_gun[df_gun['Police involvement'] == 0]['Age'], label='No Police Involvement', shade=True)
sns.kdeplot(df_gun[df_gun['Police involvement'] == 1]['Age'], label='Police Involvement', shade=True)
plt.title('Age Distribution Based on Police Involvement')
plt.xlabel('Age')
plt.ylabel('Density')
plt.legend()
plt.show()
```

```python
#Heatmap
cross_tab = pd.crosstab(df_gun['Race'], df_gun['Police involvement'])
plt.figure(figsize=(10, 6))
sns.heatmap(cross_tab, annot=True, fmt='d', cmap='coolwarm', cbar=True)
plt.title('Heatmap of Cross-Tabulation between Race and Police Involvement')
plt.xlabel('Police Involvement')
plt.ylabel('Race')
plt.show()
```

```python
# pairplot
numerical_columns = ['Age', 'Year', 'Month']

plt.figure(figsize=(15, 10))
sns.pairplot(df_gun, hue='Sex', markers=["o", "s"], vars=numerical_columns, diag_kind='kde', palette='Set2', plot_kws={'alpha':0.5})
plt.show()
```

```python
#Lineplot
numerical_x = 'Year'
numerical_y = 'Age'

plt.figure(figsize=(10, 6))
sns.lineplot(x=numerical_x, y=numerical_y, data=df_gun, marker='o', palette='Set2')
plt.xlabel('Year')
plt.ylabel('Age')
plt.show()
```

```python
#Heatmap
pivot_table = df_gun.pivot_table(values='Age', index='Month', columns='Year')

plt.figure(figsize=(12, 8))
sns.heatmap(pivot_table, cmap='viridis', annot=True, fmt=".1f", linewidths=.5)
plt.title('Heatmap of Age across Years and Months')
plt.xlabel('Year')
plt.ylabel('Month')
plt.show()
```

```python
#Clustermap
pivot_table = df_gun.pivot_table(values='Age', index='Month', columns='Year')

plt.figure(figsize=(12, 8))
sns.clustermap(pivot_table, cmap='viridis', annot=True, fmt=".1f", linewidths=.5)
plt.title('Clustermap of Age across Years and Months')
plt.xlabel('Year')
plt.ylabel('Month')
plt.show()
```
